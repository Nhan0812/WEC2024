```python
null_data = data_counties.isnull().sum()
null_percent = data_counties.isnull().sum() / data_counties.shape[0] * 100

# Create a DataFrame to display the results
null_summary = pd.DataFrame({"Null Count": null_data, "Null Percentage": null_percent})

# Sort the DataFrame in descending order based on "Null Percentage"
null_summary.sort_values(by="Null Percentage", ascending=False)
```

```python
null_data = data_muni.isnull().sum()
null_percent = data_muni.isnull().sum() / data_muni.shape[0] * 100

# Create a DataFrame to display the results
null_summary = pd.DataFrame({"Null Count": null_data, "Null Percentage": null_percent})

# Sort the DataFrame in descending order based on "Null Percentage"
null_summary.sort_values(by="Null Percentage", ascending=False)
```

```python
import plotly.express as px

# Creating a bar plot using Plotly Express
null_percent = null_percent.sort_values(ascending=True)
fig = px.bar(
    x=null_percent.index,
    y=null_percent.values,
    labels={"x": "Columns", "y": "Missing Values Percentage"},
    title="Percentage of Missing Values in Each Column",
)
fig.update_layout(width=800, height=500)
fig.show()
```

```python
# Assuming data_muni is your DataFrame with 'percent_vaccinated' column
plt.figure(figsize=(10, 6))
plt.hist(data_muni["percent_vaccinated"], bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Percent Vaccinated")
plt.xlabel("Percent Vaccinated")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
data_muni_numeric = data_muni.select_dtypes(exclude=["object"])
```

```python
# Find correlation with the "percent_vaccinated" and sort
correlations = data_muni_numeric.corr()["percent_vaccinated"].sort_values()

# Sort by most positive correlations
print(
    "Most Positive Correlations:\n", correlations.sort_values(ascending=False).head(20)
)

# Sort by most negative correlations
print("\nMost Negative Correlations:\n", correlations.sort_values().head(20))
```

```python
import statsmodels.api as sm

# Perform regression
X = sm.add_constant(data_muni["area_km2"])
y = data_muni["percent_vaccinated"]

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
data_muni["density"] = data_muni["area_km2"] / data_muni["population_total"]
```

```python
# Perform regression
X = sm.add_constant(data_muni["density"])
y = data_muni["percent_vaccinated"]

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
